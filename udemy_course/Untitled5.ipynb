```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials

scope = ['https://spreadsheets.google.com/feeds','https://www.googleapis.com/auth/drive']
```

```python
creds = ServiceAccountCredentials.from_json_keyfile_name('ex_gsheets_creds.json', scope)
```

```python
client = gspread.authorize(creds)

sheet = client.open('structure_of_sales_department')
py_gsheets = sheet.worksheet('str_sellers').get_all_records()

import pandas as pd
```

```python
df = pd.DataFrame(py_gsheets)
df[['cat_id','manager_id','post_id','rnum','name']]
```
